```python
import pandas as pd

# Load the dataset
file_path = "/mnt/data/listings_detailed.csv"
df = pd.read_csv(file_path)

# Display the first few rows to understand the structure
df.head()

# Checking relevant columns for language detection and price analysis
columns_to_check = ['name', 'description', 'neighbourhood_cleansed', 'price']
df[columns_to_check].head()

import langdetect
import numpy as np

# Function to detect language of the listing based on 'name' and 'description'
def detect_language(text):
    try:
        return langdetect.detect(text)
    except:
        return np.nan

# Applying language detection on the 'name' and 'description' fields
df['language_name'] = df['name'].astype(str).apply(detect_language)
df['language_description'] = df['description'].astype(str).apply(detect_language)

# Considering a listing to be in English if either 'name' or 'description' is in English
df['language'] = df.apply(lambda row: 'en' if 'en' in [row['language_name'], row['language_description']] else 'es', axis=1)

# Converting price to numeric values
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)

# Checking price difference between English and Spanish listings
price_comparison = df.groupby('language')['price'].agg(['mean', 'median', 'count'])

# Checking price difference across districts
district_price_comparison = df.groupby(['neighbourhood_cleansed', 'language'])['price'].agg(['mean', 'median', 'count']).unstack()

# Display the results
import ace_tools as tools

tools.display_dataframe_to_user(name="Price Comparison by Language", dataframe=price_comparison)
tools.display_dataframe_to_user(name="Price Comparison by District and Language", dataframe=district_price_comparison)

# Since langdetect is not available, using an alternative method with simple keyword-based detection

# Keywords to classify listings as English or Spanish
english_keywords = ["room", "apartment", "studio", "bathroom", "terrace", "bright", "cozy", "spacious"]
spanish_keywords = ["habitacion", "apartamento", "estudio", "baño", "terraza", "luminoso", "acogedor", "amplio"]

# Function to classify language based on keywords
def classify_language(text):
    text = str(text).lower()
    if any(word in text for word in english_keywords):
        return 'en'
    elif any(word in text for word in spanish_keywords):
        return 'es'
    return 'unknown'

# Applying language classification
df['language'] = df['name'].apply(classify_language)
df.loc[df['language'] == 'unknown', 'language'] = df['description'].apply(classify_language)

# Converting price to numeric values
df['price'] = df['price'].replace('[\$,]', '', regex=True).astype(float)

# Checking price difference between English and Spanish listings
price_comparison = df.groupby('language')['price'].agg(['mean', 'median', 'count'])

# Checking price difference across districts
district_price_comparison = df.groupby(['neighbourhood_cleansed', 'language'])['price'].agg(['mean', 'median', 'count']).unstack()

# Display the results
import ace_tools as tools

tools.display_dataframe_to_user(name="Price Comparison by Language", dataframe=price_comparison)
tools.display_dataframe_to_user(name="Price Comparison by District and Language", dataframe=district_price_comparison)

This was output-

import matplotlib.pyplot as plt
import seaborn as sns

# Set style for better visuals
sns.set_style("whitegrid")

# 1. Distribution of Prices by Listing Language
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['language'], y=df['price'], palette="coolwarm")
plt.ylim(0, 500)  # Limit outliers for better visualization
plt.title("Distribution of Prices by Listing Language", fontsize=14, fontweight='bold')
plt.xlabel("Listing Language", fontsize=12)
plt.ylabel("Price (in local currency)", fontsize=12)
plt.show()

# 2. Impact of Number of Accommodates on Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['accommodates'], y=df['price'], alpha=0.5, color="blue")
plt.title("Price vs. Number of Accommodates", fontsize=14, fontweight='bold')
plt.xlabel("Number of Accommodates", fontsize=12)
plt.ylabel("Price (in local currency)", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# 3. Average Price by Room Type
plt.figure(figsize=(8, 5))
sns.barplot(x=df['room_type'], y=df['price'], palette="viridis", estimator=np.mean)
plt.title("Average Price by Room Type", fontsize=14, fontweight='bold')
plt.xlabel("Room Type", fontsize=12)
plt.ylabel("Average Price", fontsize=12)
plt.xticks(rotation=45)
plt.show()

# 4. Average Price by District (Top 15 Expensive Districts)
district_avg_prices = df.groupby("neighbourhood_cleansed")["price"].mean().sort_values(ascending=False).head(15)
plt.figure(figsize=(12, 6))
district_avg_prices.plot(kind='bar', colormap="coolwarm", alpha=0.9)
plt.title("Top 15 Expensive Districts by Average Price", fontsize=14, fontweight='bold')
plt.xlabel("Districts", fontsize=12)
plt.ylabel("Average Price", fontsize=12)
plt.xticks(rotation=90)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

```
